```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

<h1>Coin Toss Simulation</h1>

```python
#find the heads and tails
coin = np.random.randint(2, size=10000)
```

histogram

```python
counts, bins, patches = plt.hist(coin,bins=[-0.5, 0.5, 1.5], edgecolor='black', color='skyblue')
patches[0].set_facecolor('red')    # Color for Tails
patches[1].set_facecolor('green')  # Color for Heads

plt.ylabel('Frequency')
plt.xlabel('0-Tails, 1- Heads')
plt.title('Frequency Histogram')
plt.xticks([0,1])
plt.show()
```
